```python
df = pd.read_csv("/kaggle/input/bitcoin-historical-data/bitstampUSD_1-min_data_2012-01-01_to_2021-03-31.csv")
```

```python
plt.plot(df.Timestamp, df.Open, color="blue", label="BTC Open")
plt.xlabel("Timestamp")
plt.ylabel("Open")
plt.legend()
plt.show()

plt.plot(df.Timestamp, df.Close, color="green", label="BTC Close")
plt.xlabel("Timestamp")
plt.ylabel("Close")
plt.legend()
plt.show()
```

```python
plt.scatter(df.Timestamp, df.Weighted_Price, color="red", label="price")
plt.legend()
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.title("Weighted Price")
plt.show()
```

```python
df.plot(kind="scatter", x="Volume_(BTC)", y="Volume_(Currency)",alpha = 0.5,color = "purple")
plt.xlabel("Volume_(BTC)")     
plt.ylabel("Volume_(Currency)")
plt.title("Volume Scatter Plot") 
plt.show()
```

```python
df.plot(kind="scatter",x="Open",y="Close",color="pink")
plt.xlabel("Open")
plt.ylabel("Close")
plt.title("Open-Close Scatter Plot")
plt.show()
```

```python
df.High.plot(kind = "line", color = "red",label = "High",linewidth=1,alpha =1,grid = True,linestyle = ':')
df.Low.plot(color = "yellow",label = "Low",linewidth=1, alpha = 0.5,grid = True,linestyle = '-')
plt.legend(loc="upper right")
plt.xlabel("x axis")             
plt.ylabel("y axis")
plt.title("high-low line plot")           
plt.show()
```

```python
f,ax = plt.subplots(figsize=(15, 15))
sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt= '.2f',ax=ax)
plt.show()
```
