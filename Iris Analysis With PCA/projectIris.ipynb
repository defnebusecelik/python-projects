 ## PCA

```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
url="iris.data"
data= pd.read_csv(url, names=["sepal_length","sepal_width","petal_length","petal_width","target"])
```

```python
data
```

```python
features=["sepal_length", "sepal_width","petal_length","petal_width"]
x=data[features]
y=data[["target"]]
```

```python
#standardizing
x=StandardScaler().fit_transform(x)
```

```python
x
```

```python
#PCA
pca=PCA(n_components=2) #2 dimension
principalComponents=pca.fit_transform(x)
principalData=pd.DataFrame(data=principalComponents, columns=["principal_component_1","principal_component_2"])
```

```python
principalData
```

```python
finalData=pd.concat([principalData, data[["target"]]],axis=1)
```

```python
finalData.head()
```

```python
#visualization
dataSetosa=finalData[data.target=="Iris-setosa"]
dataVirginica=finalData[data.target=="Iris-virginica"]
dataVersicolor=finalData[data.target=="Iris-versicolor"]

plt.xlabel("principal component 1")
plt.ylabel("principal component 2")

plt.scatter(dataSetosa["principal_component_1"], dataSetosa["principal_component_2"], color="purple")
plt.scatter(dataVirginica["principal_component_1"], dataVirginica["principal_component_2"], color="blue")
plt.scatter(dataVersicolor["principal_component_1"],dataVersicolor["principal_component_2"], color="pink")
```
